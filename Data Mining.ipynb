```python
def get_posts(user: str, limit=None):
    links = []
    for num in range(9**999):
        url = f"https://habr.com/ru/users/{user}/posts/page{num}"
        request = requests.get(url)
        soup = BeautifulSoup(request.text)
        posts = soup.find_all("li", "content-list__item content-list__item_post shortcuts_item")
        
        if len(posts):
            posts = [post.find("a", "post__title_link").get("href") for post in posts]
            links.extend(posts)
            
            if limit and len(links) >= limit:
                return links[:limit]
        else:
            break
            
    return links
```

```python
def get_text(link: str):
    request = requests.get(link)
    if request.status_code != 200:
        return ""
    text = request.text
    soup = BeautifulSoup(text)
    text_post = soup.find("div", "post__text", id="post-content-body").text
    return text_post
```

```python
def get_avg_length_word(tokens: list):
    return sum(len(token) for token in tokens) / len(tokens)
```

```python
from razdel import sentenize, tokenize
```

```python
def get_tokens(text: str, lower=False, delete_punctuations=False) -> list:
    if lower:
        text = text.lower()
    tokens = [token.text for token in tokenize(text)]
    
    if delete_punctuations:
        tokens = [token for token in tokens if token not in ".,?!-\";:()—"]
    
    return tokens
```

```python
def get_sentences(text: str) -> list:
    return [sentence.text for sentence in sentenize(text)]
```

```python
def get_avg_length_sentences(sentences: list):
    return sum(len(sentence) for sentence in sentences) / len(sentences)
```

```python
def get_avg_count_words_per_sentence(sentences: list):
    count_words = 0
    for sentence in sentences:
        tokens = get_tokens(sentence, lower=True, delete_punctuations=True)
        count_words += len(tokens)
    
    return count_words / len(sentences)
```

```python
def get_distribution_pos(tokens: list) -> list:
    morph = pymorphy2.MorphAnalyzer()
    distribution = dict.fromkeys(list(morph.TagClass.PARTS_OF_SPEECH), 0)
    
    for token in tokens:
        pos = morph.parse(token)[0].tag.POS
        if pos in distribution:
            distribution[pos] += 1
    
    summary = sum(distribution.values())
    distribution = {first: second / summary for first, second in distribution.items()}
    
    return distribution
```

```python
def process_data(author, text):
    data = []
    # Remove the space symbols
    text = re.sub(r"[\s]+", r" ", text).strip()
    
    # Get sentences from the text
    sentences = get_sentences(text)
    # Get tokens from the text with punctuations
    tokens = get_tokens(text, lower=True)
    # feature 1
    avg_cnt_puncts = len(get_punctuations(tokens)) / len(tokens)
    
    # Get tokens from the text without punctuations
    tokens = get_tokens(text, lower=True, delete_punctuations=True)
    # feature 2
    avg_len_word = get_avg_length_word(tokens)
    # feature 3
    avg_cnt_words = get_avg_count_words_per_sentence(sentences)
    # feature 4
    avg_len_sentences = get_avg_length_sentences(sentences)
    # feature 5
    distributions = get_distribution_pos(tokens)
    # feature 6
    avg_foreign_words = len(get_foreign_words(tokens))/len(tokens)
    
    data = [avg_cnt_puncts, avg_len_word, avg_cnt_words, avg_len_sentences, avg_foreign_words, *list(distributions.values())]
    columns = ['author', 'text', 'avg_cnt_puncts', 'avg_len_word', 'avg_cnt_words', 'avg_len_sentences', 'avg_foreign_words']
    columns.extend(list(distributions.keys()))
    
    return columns, data
```
