```python
def get_digits_from_text(text: str):
    return re.findall(r"\s\d+\s", text)
    
def get_ranges_from_text(text: str):
    return re.findall(r"[\d]+-[\d]+", text)

def get_serial_digits(text: str):
    return re.findall(r"[\d]+-[^\d]+")
```

```python
data = defaultdict(list)
for user in USERS:
    for num in range(9**999):
        url = f"https://habr.com/ru/users/{user}/posts/page{num}"
        request = requests.get(url)
        soup = BeautifulSoup(request.text)
        posts = soup.find_all("li", "content-list__item content-list__item_post shortcuts_item")
        
        if len(posts):
            posts = [post.find("a", "post__title_link").get("href") for post in posts]
            data[user].extend(posts)
        else:
            break
```

```python
def get_posts(user: str):
    links = []
    for num in range(9**999):
        url = f"https://habr.com/ru/users/{user}/posts/page{num}"
        request = requests.get(url)
        soup = BeautifulSoup(request.text)
        posts = soup.find_all("li", "content-list__item content-list__item_post shortcuts_item")
        
        if len(posts):
            posts = [post.find("a", "post__title_link").get("href") for post in posts]
            links.extend(posts)
        else:
            break
            
    return links
```

```python
def get_text(link: str):
    request = requests.get(link)
    if request.status_code != 200:
        return ""
    text = request.text
    soup = BeautifulSoup(text)
    text_post = soup.find("div", "post__text", id="post-content-body").text
    return text_post
```

```python
get_text('https://habr.com/ru/news/t/384429/')
```
